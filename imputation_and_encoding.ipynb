```python
import pandas as pd
import numpy as np
```

```python
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.isna().sum()
```

```python
car_sales_missing.dropna(subset = "Price" , inplace = True)
```

```python
car_sales_missing.isna().sum()
```

```python
X = car_sales_missing.drop(labels = "Price" , axis = 1) 
y = car_sales_missing["Price"]
X.isna().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

catogory_impute = SimpleImputer(strategy = "constant" , fill_value = "missing")
door_impute = SimpleImputer(strategy = "constant" , fill_value = 4)
odom_impute = SimpleImputer(strategy = "mean")

catogory_name = ["Make" , "Colour"]
door_name = ["Doors"]
odom_name = ["Odometer (KM)"]

transform = ColumnTransformer( transformers= [
    ("cat" , catogory_impute , carogory_name) ,
                             ("door" , door_impute , door_name ) ,
                             ("odom", odom_impute , odom_name)])
filled = transform.fit_transform(X)
filled
```

```python
filled_df = pd.DataFrame(filled)
```

```python
filled_df
```

```python
filled_df.head(14)
```

## now lets convert the names or catogory into numbers so the computer can understand

```python
filled_df.rename(columns = { 0 : "Make" , 1 : "Colour" , 2 : "Door" , 3 : "odometer"} , inplace = True)
```

```python
X = filled_df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

catagory_features = ["Make" , "Colour" , "Door"]
one_hot = OneHotEncoder(sparse_output= False)
transform = ColumnTransformer([("covert_into_num" , one_hot , catagory_features) ] , remainder = 'passthrough')
transformed = transform.fit_transform(X)
```
